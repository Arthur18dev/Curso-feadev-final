## **📊 Descrição das Colunas do DataFrame**

### **Identificação do Estudante**  
- **ID_Aluno**: Identificador único para cada aluno (de 1001 a 3392).

### **👥 Detalhes Demográficos**  
- **Idade**: Idade dos estudantes, variando de 15 a 18 anos.  
- **Genero**: Gênero do estudante:
  - 0: Masculino
  - 1: Feminino  
- **Etnia**: Etnia dos estudantes:
  - 0: Caucasiano
  - 1: Afrodescendente
  - 2: Asiático
  - 3: Outra

### **👨‍👩‍👧 Educação dos Pais**  
- **Escolaridade_Pais**: Nível de escolaridade dos pais:
  - 0: Nenhum
  - 1: Ensino Médio
  - 2: Alguma Faculdade
  - 3: Graduação
  - 4: Pós-Graduação ou Superior

### **📚 Hábitos de Estudo**  
- **Horas_Estudo_Semanais**: Tempo de estudo semanal em horas (0 a 20).  
- **Faltas**: Número de faltas no ano letivo (0 a 30).  
- **Aulas_Particulares**: Participação em aulas particulares:
  - 0: Não
  - 1: Sim

### **👨‍👩‍👧 Envolvimento Parental**  
- **Apoio_Pais**: Nível de apoio dos pais:
  - 0: Nenhum
  - 1: Baixo
  - 2: Moderado
  - 3: Alto
  - 4: Muito Alto

### **🏫 Atividades Extracurriculares**  
- **Atividades_Extracurriculares**: Participação em atividades extracurriculares:
  - 0: Não
  - 1: Sim  
- **Esportes**: Participação em esportes:
  - 0: Não
  - 1: Sim  
- **Musica**: Participação em atividades musicais:
  - 0: Não
  - 1: Sim  
- **Voluntariado**: Participação em atividades de voluntariado:
  - 0: Não
  - 1: Sim

### **📈 Desempenho Acadêmico**  
- **Media_Notas**: Média das notas (GPA), variando de 2.0 a 4.0.  
- **Classificacao_Notas**: Classificação do aluno com base na média das notas:
  - 0: 'A' (Média >= 3.5)
  - 1: 'B' (3.0 <= Média < 3.5)
  - 2: 'C' (2.5 <= Média < 3.0)
  - 3: 'D' (2.0 <= Média < 2.5)
  - 4: 'F' (Média < 2.0)

```python
# Importando as bibliotecas estudadas durante o curso
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Importando o dataset

# Verificando as primeiras linhas do dataset
```

```python
data = {
    "ID_Aluno": [1, 2, 3, 4, 5],
    "Idade": [16, 17, 16, 18, 17],
    "Genero": ['F', 'M', 'M', 'F', 'M'],
    "Etnia": ['Branca', 'Negra', 'Parda', 'Branca', 'Negra'],
    "Escolaridade_Pais": ['Superior', 'Ensino Médio', 'Ensino Médio', 'Superior', 'Ensino Fundamental'],
    "Horas_Estudo_Semanais": [10, 5, 8, 6, 12],
    "Faltas": [2, 5, 0, 3, 1],
    "Aulas_Particulares": [1, 0, 1, 0, 1],
    "Apoio_Pais": [1, 0, 1, 1, 0],
    "Atividades_Extracurriculares": [1, 1, 0, 1, 0],
    "Esportes": [1, 0, 1, 0, 1],
    "Musica": [0, 1, 0, 1, 1],
    "Voluntariado": [0, 1, 1, 0, 0],
    "Media_Notas": [7.5, 6.0, 8.0, 7.0, 9.0],
    "Classificacao_Notas": ['B', 'C', 'B', 'B', 'A']
}

df = pd.DataFrame(data)

```

```python
# Agora está por conta própria e pronto para brilhar! 🌟

### 🔍 **Pergunta Norteadora** ###
# Quais fatores demográficos, acadêmicos e comportamentais mais influenciam a classificação e a média de notas dos estudantes?

# 🛠️ **Dicas para começar:**
# - Dê uma boa olhada no dataset.
# - Explore as relações entre a variável `Classificacao_Notas` e outros fatores, como gênero, horas de estudo, faltas, etc.
# - Use gráficos e tabelas para visualizar os padrões encontrados.
# - Escreva suas reflexões e hipóteses sobre o que você encontrou.
# - E o mais importante: divirta-se no processo de análise! 😄

# Não há limite para o número de células ou análises. O espaço é todo seu! 🚀

# **SEU CÓDIGO COMEÇA AQUI** 👇
classificacao_map = {'A': 4, 'B': 3, 'C': 2, 'D': 1}
df['Classificacao_Notas_num'] = df['Classificacao_Notas'].map(classificacao_map)

correlacao = df[['Idade', 'Horas_Estudo_Semanais', 'Faltas', 'Media_Notas', 'Classificacao_Notas_num']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlação entre Variáveis")
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Classificacao_Notas', hue='Genero')
plt.title('Classificação das Notas por Gênero')
plt.xlabel('Classificação das Notas')
plt.ylabel('Contagem')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Atividades_Extracurriculares', y='Classificacao_Notas_num')
plt.title('Distribuição da Classificação das Notas por Atividades Extracurriculares')
plt.xlabel('Atividades Extracurriculares (1 = Participa, 0 = Não Participa)')
plt.ylabel('Classificação das Notas')
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Horas_Estudo_Semanais', y='Classificacao_Notas_num', hue='Genero', palette='coolwarm')
plt.title('Horas de Estudo vs. Classificação das Notas')
plt.xlabel('Horas de Estudo Semanais')
plt.ylabel('Classificação das Notas')
plt.show()

```
